```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('spam.csv', encoding='latin1')
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
```

```python
X = df['v2']
y = df['v1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

```python
predictions = model.predict(X_test_tfidf)
print(classification_report(y_test, predictions))
```

```python
def predict_spam(message):
    # Vectorize the message
    message_tfidf = vectorizer.transform([message])
    # Predict using the trained model
    prediction = model.predict(message_tfidf)
    return prediction[0]

# Create a list of messages to test
test_messages = [
    "Congratulations! You've won a free prize, click here to claim it!",
    "Hey, are we still meeting for lunch today?",
    "URGENT: Your account has been compromised, verify now!",
    "Can you send me the report by EOD?"
]

# Loop through the list and print the results
print("-- Model Predictions --")
for msg in test_messages:
    result = predict_spam(msg)
    print(f"Message: '{msg}'\nPrediction: {result}\n")
```

## Project Summary: Spam Detection with Naive Bayes

This notebook demonstrates a complete workflow for building a spam detection system using a Multinomial Naive Bayes classifier. The key steps performed are:

1.  **Data Loading and Initial Inspection**: The `spam.csv` dataset was loaded into a pandas DataFrame. An initial `UnicodeDecodeError` was resolved by specifying `encoding='latin1'`.

2.  **Data Preparation**: The relevant columns were identified and extracted: 'v2' containing the message text (`X`) and 'v1' containing the spam/ham labels (`y`).

3.  **Train-Test Split**: The dataset was split into training and testing sets to evaluate the model's generalization performance.

4.  **Text Vectorization (TF-IDF)**: `TfidfVectorizer` was used to convert the raw text messages into numerical feature vectors. This technique accounts for the importance of words within documents and across the entire corpus, with 'english' stop words removed.

5.  **Model Training**: A `MultinomialNB` classifier was trained on the TF-IDF transformed training data and labels.

6.  **Model Evaluation**: The model's performance was assessed using a `classification_report` on the test set. The report provides precision, recall, f1-score, and support for both 'ham' and 'spam' classes, demonstrating high accuracy.

7.  **Custom Prediction Function**: A `predict_spam` function was created to allow for easy testing of new, unseen messages. This function preprocesses the input message using the trained vectorizer and then makes a prediction using the trained model. This function was successfully demonstrated with several example messages.
